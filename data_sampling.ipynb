```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = '202311-citibike-tripdata.csv'
df = pd.read_csv('data/'+filename)

display(df)
```

```python
# checking for null
print(df.isna().sum())

print('\n\ndataframe shape')
print(df.shape)
```

```python
# dropping null values
df.dropna(axis = 0, inplace=True)

print(df.isna().sum())
print('\n\nUpdated Dataframe Shape after dropping columns')
print(df.shape)
```

```python
# EDA
print('Dataframe Information')
print(df.info())
```

```python
df.head()
```

```python
# Converting Datatypes
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['ride_id'] = df['ride_id'].astype('str')
df['rideable_type'] = df['rideable_type'].astype('str')
df['start_station_name'] = df['start_station_name'].astype('str')
df['start_station_id'] = df['start_station_id'].astype('str')
df['end_station_name'] = df['end_station_name'].astype('str')
df['end_station_id'] = df['end_station_id'].astype('str')
df['member_casual'] = df['member_casual'].astype('str')

print('Dataframe Information after Datatype conversion')
print(df.info())
```

```python
temporal_interval = 'H'

sampled_data = df.groupby(pd.Grouper(key='started_at', freq=temporal_interval)).apply(lambda x: x.sample(n=1, random_state=42) if not x.empty else None)

# Filter out empty groups

sampled_data = sampled_data.dropna().reset_index(drop=True)

display(sampled_data)
print(sampled_data.info())
```

```python
print('Original Dataframe Counts')
value_counts = df['rideable_type'].value_counts()
# value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.show()
print('Ratio Classic to Electric',value_counts['classic_bike']/value_counts['electric_bike'])



value_counts = df['member_casual'].value_counts()
# value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.show()
print('Ratio Member to Casuals',value_counts['member']/value_counts['casual'])
```

```python
print('Sampled Dataframe Counts')
value_counts = sampled_data['rideable_type'].value_counts()
# value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.show()
print('Ratio Classic to Electric',value_counts['classic_bike']/value_counts['electric_bike'])

value_counts = sampled_data['member_casual'].value_counts()
# value_counts.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.show()
print('Ratio Member to Casuals',value_counts['member']/value_counts['casual'])
```

```python
sampled_data.to_csv('data/sampled/'+filename)
```
